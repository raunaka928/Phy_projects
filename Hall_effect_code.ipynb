```python
import numpy as np

import matplotlib.pyplot as plt

import pandas as pd

from scipy import stats, constants, optimize
```

# 300K Constant 70G Pins 2-3


```python
def func(x, a, b):
    return a * x + b

# Read data from CSV file
df_sheet300k = pd.read_csv('300k_0.070kG_2-3.csv')

# Assume the CSV file has columns 'x' and 'y'
x_sheet300k = df_sheet300k['Current uA'] * constants.micro
y_sheet300k = df_sheet300k['Voltage mV'] * constants.milli

# Use curve_fit to fit the function to your data, getting the optimal parameters
popt, pcov = optimize.curve_fit(func, x_sheet300k, y_sheet300k)


# Print the slope, intercept, and their respective errors
print(f"Slope: {popt[0]} ± {np.sqrt(pcov[0, 0])}")
print(f"Intercept: {popt[1]} ± {np.sqrt(pcov[1, 1])}")


print("Intercept: " + str(popt[1]) + " ± " + str(np.sqrt(pcov[1, 1])))
print("Intercept: {} ± {}".format(popt[1], np.sqrt(pcov[1, 1])))


# Create a function for the fitted line
line = popt[0] * x_sheet300k + popt[1]

print(pcov)
# Plot the data along with the fitted line
plt.figure(figsize=(10, 6))
plt.plot(x_sheet300k, y_sheet300k, 'o', label='original data')
plt.plot(x_sheet300k, line, 'r', label='fitted line')

# Add x and y labels
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()
plt.show()
```

# 300K Constant 1.542kG Pins 2-3

```python
# Read data from CSV file
df_magneto300k = pd.read_csv('300k_1.542kG_2-3.csv')

# Assume the CSV file has columns 'x' and 'y'
x_magneto300k = df_magneto300k['Current uA'] * constants.micro
y_magneto300k = df_magneto300k['Voltage mV'] * constants.milli

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(x_magneto300k, y_magneto300k)

print(slope, intercept, r_value, p_value, std_err)

# Create a function for the linear regression line
line = slope * x_magneto300k + intercept

# Plot the data along with the fitted line
plt.figure(figsize=(10, 6))
plt.plot(x_magneto300k, y_magneto300k, 'o', label='original data')
plt.plot(x_magneto300k, line, 'r', label='fitted line')
#plt.errorbar(x, y, yerr=std_err, fmt='o', label='error bars')
plt.legend()
plt.show()
```
